## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor

# For reproducibility
np.random.seed(42)

# ----------------------------
# 1. Simulate Advertisement Data
# ----------------------------
N = 1000  # number of observations

# Simulate customer features:
# - 'age' between 18 and 70
# - 'income' (in dollars) normally distributed
# - 'prior_clicks' (a proxy for engagement) from a Poisson distribution
age = np.random.randint(18, 70, size=N)
income = np.random.normal(50000, 10000, size=N)
prior_clicks = np.random.poisson(5, size=N)

# Simulate treatment assignment (advertisement exposure)
# Assume a randomized experiment: treatment probability = 0.5
treatment = np.random.binomial(1, 0.5, size=N)

# Create a baseline outcome.
# Imagine the outcome is something like revenue or engagement score.
# For control group (no ad exposure), outcome depends on age and income.
baseline = 20 + 0.3 * age + 0.5 * (income / 1000)  # scale income for simplicity

# Define a heterogeneous treatment effect (the "true" CATE)
# Here, the treatment effect increases with age and decreases with prior clicks.
true_effect = 5 + 0.1 * age - 0.2 * prior_clicks

# Add noise to the outcome
noise = np.random.normal(0, 5, size=N)

# Simulate the observed outcome:
# For the treated, the outcome is the sum of the baseline, the heterogeneous treatment effect, and noise.
# For the control group, only the baseline + noise matters.
outcome = baseline + treatment * true_effect + noise

# Create a DataFrame with our simulated data
data = pd.DataFrame({
    'age': age,
    'income': income,
    'prior_clicks': prior_clicks,
    'treatment': treatment,
    'outcome': outcome,
    'true_effect': true_effect  # saving the true effect for evaluation
})

print("First five rows of the simulated dataset:")
print(data.head())

# ----------------------------
# 2. Estimate CATE using a T-Learner Approach
# ----------------------------
# In a T-learner, we fit one model for the treated and another for the controls.
# Here we use RandomForestRegressor for flexibility and to capture non-linearities.

# Define the feature set
features = ['age', 'income', 'prior_clicks']
X = data[features]

# Split the data by treatment group
treated = data[data.treatment == 1]
control = data[data.treatment == 0]

# Fit a model on the treated group
model_treated = RandomForestRegressor(n_estimators=100, random_state=42)
model_treated.fit(X.loc[treated.index], treated.outcome)

# Fit a model on the control group
model_control = RandomForestRegressor(n_estimators=100, random_state=42)
model_control.fit(X.loc[control.index], control.outcome)

# For every observation, predict the outcome under treatment and under control.
data['y1_hat'] = model_treated.predict(X)
data['y0_hat'] = model_control.predict(X)

# The estimated CATE for each observation is the difference in these predictions.
data['cate_est'] = data['y1_hat'] - data['y0_hat']

print("\nFirst five rows with estimated treatment effects:")
print(data[['age', 'income', 'prior_clicks', 'treatment', 'outcome', 'cate_est', 'true_effect']].head())

# ----------------------------
# 3. Visualizations
# ----------------------------

# (A) Scatter Plot: True vs. Estimated Treatment Effect
plt.figure(figsize=(10, 6))
sns.scatterplot(x='true_effect', y='cate_est', data=data, alpha=0.6)
plt.xlabel("True Treatment Effect")
plt.ylabel("Estimated Treatment Effect")
plt.title("True vs. Estimated Conditional Average Treatment Effects")
# Plot a red dashed line for the ideal 45 reference line
min_effect = min(data.true_effect.min(), data.cate_est.min())
max_effect = max(data.true_effect.max(), data.cate_est.max())
plt.plot([min_effect, max_effect], [min_effect, max_effect], color='red', linestyle='--', label='Ideal')
plt.legend()
plt.show()

# (B) Histogram of the Estimated Treatment Effects
plt.figure(figsize=(10, 6))
sns.histplot(data['cate_est'], kde=True, color='skyblue')
plt.xlabel("Estimated Treatment Effect")
plt.title("Distribution of Estimated Conditional Average Treatment Effects")
plt.show()

# (C) Bar Plot: Estimated Treatment Effect by Age Group
# Bin age into 10 groups and compute the average estimated treatment effect in each bin.
data['age_bin'] = pd.cut(data['age'], bins=10)
age_effect = data.groupby('age_bin')['cate_est'].mean().reset_index()

plt.figure(figsize=(12, 6))
sns.barplot(x='age_bin', y='cate_est', data=age_effect, palette='viridis')
plt.xlabel("Age Bin")
plt.ylabel("Average Estimated Treatment Effect")
plt.title("Estimated Treatment Effect by Age Group")
plt.xticks(rotation=45)
plt.show()

# (D) Scatter with LOESS smoothing: Estimated Treatment Effect vs. Age
plt.figure(figsize=(12, 6))
sns.regplot(x='age', y='cate_est', data=data, lowess=True, scatter_kws={'alpha': 0.5}, line_kws={'color': 'red'})
plt.xlabel("Age")
plt.ylabel("Estimated Treatment Effect")
plt.title("Estimated Treatment Effect vs. Age with LOESS Smoothing")
plt.show()
```
