## Data science
</div>

With Colab you can harness the full power of popular Python libraries to analyse and visualise data. The code cell below uses <strong>numpy</strong> to generate some random data, and uses <strong>matplotlib</strong> to visualise it. To edit the code, just click the cell and start editing.

```
import numpy as np
import IPython.display as display
from matplotlib import pyplot as plt
import io
import base64

ys = 200 + np.random.randn(100)
x = [x for x in range(len(ys))]

fig = plt.figure(figsize=(4, 3), facecolor='w')
plt.plot(x, ys, '-')
plt.fill_between(x, ys, 195, where=(ys > 195), facecolor='g', alpha=0.6)
plt.title("Sample Visualization", fontsize=10)

data = io.BytesIO()
plt.savefig(data)
image = F"data:image/png;base64,{base64.b64encode(data.getvalue()).decode()}"
alt = "Sample Visualization"
display.display(display.Markdown(F"""![{alt}]({image})"""))
plt.close(fig)
```

```
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Root node representing the overall skillset
root = "Business Analyst Skills"
G.add_node(root)

# Main categories
business_knowledge = "Business Knowledge"
data_proficiency = "Data Proficiency"
G.add_nodes_from([business_knowledge, data_proficiency])
G.add_edge(root, business_knowledge)
G.add_edge(root, data_proficiency)

# Subskills for Business Knowledge
business_skills = [
    "Operations",
    "Marketing",
    "Business Strategy",
    "Economics"
]
for skill in business_skills:
    G.add_node(skill)
    G.add_edge(business_knowledge, skill)

# Subskills for Data Proficiency
data_skills = [
    "Data Analysis Tools (Excel/Python)",
    "Visualization Platforms (Tableau, Power BI)",
    "Database Knowledge (SQL)"
]
for skill in data_skills:
    G.add_node(skill)
    G.add_edge(data_proficiency, skill)

# Attempt to use Graphviz layout for a neat hierarchical structure;
# fallback to spring_layout if Graphviz is not available.
try:
    pos = nx.nx_agraph.graphviz_layout(G, prog="dot")
except Exception as e:
    print("Graphviz layout not found, using spring_layout instead.")
    pos = nx.spring_layout(G, seed=42)

# Define custom node colors for clarity
node_colors = []
for node in G.nodes():
    if node == root:
        node_colors.append("gold")
    elif node == business_knowledge:
        node_colors.append("skyblue")
    elif node == data_proficiency:
        node_colors.append("lightgreen")
    elif node in business_skills:
        node_colors.append("lightblue")
    elif node in data_skills:
        node_colors.append("palegreen")
    else:
        node_colors.append("gray")

# Draw the graph
plt.figure(figsize=(10, 8))
nx.draw(
    G, pos,
    with_labels=True,
    node_color=node_colors,
    node_size=3000,
    font_size=10,
    font_weight="bold",
    arrows=True,
    arrowstyle='->',
    arrowsize=15
)

plt.title("Key Skills for Business Analyst Role in India")
plt.axis('off')
plt.tight_layout()
plt.show()
```

```
import chartify
import pandas as pd

# Define the skills and their grouping
data = {
    "skill": [
        "Operations", "Marketing", "Business Strategy", "Economics",
        "Excel", "Python", "Tableau", "Power BI", "SQL"
    ],
    "category": ["Business Knowledge"] * 4 + ["Data Proficiency"] * 5,
    # Dummy numeric value (all 1) used solely for plotting
    "dummy": [1] * 9
}

# Create a DataFrame and enforce the desired order of skills
df = pd.DataFrame(data)
skill_order = ["Operations", "Marketing", "Business Strategy", "Economics",
               "Excel", "Python", "Tableau", "Power BI", "SQL"]
df["skill"] = pd.Categorical(df["skill"], categories=skill_order, ordered=True)

# Define custom colors for each category
business_color = "#2E86C1"  # Blue for Business Knowledge
data_color = "#E74C3C"      # Red for Data Proficiency

# Create the Chartify chart with a categorical x-axis (skills) and a linear y-axis
chart = chartify.Chart(blank_labels=True, x_axis_type="categorical", y_axis_type="linear")
chart.set_title("Jio Marketing students - training for Business analysts")

# Plot a horizontal bar chart.
# The dummy numeric column is used to draw bars; the actual focus is on the color grouping.
chart.plot.barh(
    data_frame=df,
    categorical_columns=["skill"],
    numeric_column="dummy",
    color_column="category"
)

# Set the series colors so that each category uses our defined palette.
chart.set_series_colors([business_color, data_color])

# Hide the x-axis since the dummy values are not meaningful
chart.axes.xaxis.visible = False

# Optionally adjust the y-axis label
chart.axes.yaxis.axis_label = "Skill"

chart.show()
```

```
import matplotlib.pyplot as plt
import numpy as np

# Use a clean style
plt.style.use('seaborn-whitegrid')

# Define categories and their respective sections
business_knowledge = ["Operations", "Marketing", "Business Strategy", "Economics"]
data_proficiency = ["Excel", "Python", "Tableau", "Power BI", "SQL"]

# Define colors for categories
business_knowledge_color = "#2E86C1"  # Blue for Business Knowledge
data_proficiency_color = "#E74C3C"      # Red for Data Proficiency

# Define figure
fig, ax = plt.subplots(figsize=(12, 8))

# Global title
ax.set_title("Jio Marketing students - training for Business analysts",
             fontsize=22, fontweight='bold', color="#34495e", pad=20)

# Define bar positions with spacing between groups
offset = 1  # Extra spacing between the two groups
y_positions_business = np.arange(len(business_knowledge))
y_positions_data = np.arange(len(data_proficiency)) + len(business_knowledge) + offset

# Plot Business Knowledge bars
ax.barh(y_positions_business, [1] * len(business_knowledge),
        color=business_knowledge_color, alpha=0.85, edgecolor="none")
for i, label in enumerate(business_knowledge):
    ax.text(0.5, y_positions_business[i], label, ha='center', va='center',
            fontsize=14, fontweight='bold', color="white")

# Plot Data Proficiency bars
ax.barh(y_positions_data, [1] * len(data_proficiency),
        color=data_proficiency_color, alpha=0.85, edgecolor="none")
for i, label in enumerate(data_proficiency):
    ax.text(0.5, y_positions_data[i], label, ha='center', va='center',
            fontsize=14, fontweight='bold', color="white")

# Remove axes and ticks for a cleaner look
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

# Add section labels on the left side
ax.text(-0.05, np.mean(y_positions_business), "Business Knowledge",
        fontsize=16, fontweight='bold', color=business_knowledge_color,
        va='center', ha='right', rotation=90)
ax.text(-0.05, np.mean(y_positions_data), "Data Proficiency",
        fontsize=16, fontweight='bold', color=data_proficiency_color,
        va='center', ha='right', rotation=90)

# Adjust the x-limits to give extra space for the group labels on the left
ax.set_xlim(-0.2, 1.1)

plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt
import numpy as np
import textwrap

# Define categories and their respective skills
business_knowledge = ["Operations", "Marketing", "Business Strategy", "Economics"]
data_proficiency = ["Excel", "Python", "Tableau", "Power BI", "SQL"]

# Define course details for each skill
course_descriptions = {
    "Operations": "Operations Management course covers this",
    "Marketing": "Students are equipped (Marketing Course)",
    "Business Strategy": "Business Strategy Course course covers this",
    "Economics": "Economics for Managers course covers this",
    "Excel": "Taught Excel for Beginners; upcoming Excel training this quarter",
    "Python": "Taught Python coding via LLMs in Marketing Research Methods (Github sample)",
    "Tableau": "Will be done along with the excel",
    "Power BI": "Course planned after Excel & Tableau courses - we can also have course-era module suggested",
    "SQL": "Faculty is planned for short course (e.g., Manoranjan Pradhan): Business Analytics basics to be introduced first, then Database Basics & SQL Lab"
}

# Define colors for each category
business_knowledge_color = "#2E86C1"  # Blue
data_proficiency_color = "#E74C3C"      # Red

# Create figure and axis
fig, ax = plt.subplots(figsize=(12, 8))

# Set global title
ax.set_title("Jio Marketing students - training for Business analysts",
             fontsize=22, fontweight='bold', color="#34495e", pad=20)

# Define bar positions with extra spacing between groups
offset = 1  # Extra spacing between the two groups
y_positions_business = np.arange(len(business_knowledge))
y_positions_data = np.arange(len(data_proficiency)) + len(business_knowledge) + offset

# Plot Business Knowledge bars
ax.barh(y_positions_business, [1] * len(business_knowledge),
        color=business_knowledge_color, alpha=0.85, edgecolor="none")
for i, skill in enumerate(business_knowledge):
    # Skill name centered in the bar
    ax.text(0.5, y_positions_business[i], skill, ha='center', va='center',
            fontsize=14, fontweight='bold', color="white")
    # Course details to the right of the bar (wrapped for readability)
    desc = course_descriptions[skill]
    wrapped_desc = "\n".join(textwrap.wrap(desc, width=40))
    ax.text(1.05, y_positions_business[i], wrapped_desc, ha='left', va='center',
            fontsize=10, color="#34495e")

# Plot Data Proficiency bars
ax.barh(y_positions_data, [1] * len(data_proficiency),
        color=data_proficiency_color, alpha=0.85, edgecolor="none")
for i, skill in enumerate(data_proficiency):
    # Skill name centered in the bar
    ax.text(0.5, y_positions_data[i], skill, ha='center', va='center',
            fontsize=14, fontweight='bold', color="white")
    # Course details to the right of the bar
    desc = course_descriptions[skill]
    wrapped_desc = "\n".join(textwrap.wrap(desc, width=40))
    ax.text(1.05, y_positions_data[i], wrapped_desc, ha='left', va='center',
            fontsize=10, color="#34495e")

# Remove axes ticks and spines for a cleaner look
ax.set_xticks([])
ax.set_yticks([])
for spine in ax.spines.values():
    spine.set_visible(False)

# Add vertical section labels on the left
ax.text(-0.15, np.mean(y_positions_business), "Business Knowledge",
        fontsize=16, fontweight='bold', color=business_knowledge_color,
        va='center', ha='right', rotation=90)
ax.text(-0.15, np.mean(y_positions_data), "Data Proficiency",
        fontsize=16, fontweight='bold', color=data_proficiency_color,
        va='center', ha='right', rotation=90)

# Adjust x-limits to provide space for the course descriptions and section labels
ax.set_xlim(-0.3, 2.5)

plt.tight_layout()
plt.show()
```
