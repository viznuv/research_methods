```
import numpy as np
import matplotlib.pyplot as plt
import random

# -----------------------------
# 1. Define the Environment
# -----------------------------
class SeasonalPricingEnv:
    """
    A simple environment to simulate dynamic pricing with a seasonal demand.
    - State: time step t in [0, T_max-1]
    - Action: price chosen from a discrete set
    - Reward: price * demand
    - Demand is a function of time and price, plus random noise.
    """
    def __init__(self, T_max=10, price_set=[5, 6, 7, 8, 9]):
        self.T_max = T_max
        self.price_set = price_set
        self.num_actions = len(price_set)
        self.state = 0  # time index
        self.done = False

    def reset(self):
        """
        Reset the environment for a new episode.
        Returns the initial state.
        """
        self.state = 0
        self.done = False
        return self.state

    def step(self, action_idx):
        """
        Take one step in the environment given an action index.

        action_idx: index in the price_set.
        Returns:
          next_state, reward, done
        """
        price = self.price_set[action_idx]

        # Simulate demand based on "seasonality"
        # For simplicity, let's split the T_max into two seasons:
        #   Early season (t < T_max/2): higher base demand
        #   Late season  (t >= T_max/2): lower base demand
        # Then let the price reduce demand.
        if self.state < self.T_max // 2:
            # High demand season
            base_demand = 100
        else:
            # Low demand season
            base_demand = 60

        # Demand decreases with price, plus some noise
        demand_mean = max(base_demand - 5 * price, 0)
        demand = np.random.poisson(lam=max(demand_mean, 0))

        # Revenue = price * demand
        reward = price * demand

        # Move to next time step
        self.state += 1
        if self.state >= self.T_max:
            self.done = True

        return self.state, reward, self.done


# -----------------------------
# 2. Q-learning parameters
# -----------------------------
NUM_EPISODES = 2000      # Number of training episodes
T_MAX = 10               # Episode length
PRICE_SET = [5, 6, 7, 8, 9]
ALPHA = 0.1              # Learning rate
GAMMA = 0.95             # Discount factor
EPSILON_START = 1.0      # Initial exploration probability
EPSILON_END = 0.01       # Final exploration probability
EPSILON_DECAY = 0.999    # Decay rate per episode

# Create environment
env = SeasonalPricingEnv(T_max=T_MAX, price_set=PRICE_SET)

# Q-table shape: [T_max, number_of_actions]
Q = np.zeros((T_MAX, env.num_actions))

# Tracking
episode_rewards = []

# -----------------------------
# 3. Q-learning loop
# -----------------------------
epsilon = EPSILON_START

for episode in range(NUM_EPISODES):
    state = env.reset()
    total_reward = 0

    for t in range(T_MAX):
        # Epsilon-greedy action selection
        if random.random() < epsilon:
            action_idx = np.random.randint(env.num_actions)
        else:
            action_idx = np.argmax(Q[state, :])

        next_state, reward, done = env.step(action_idx)

        # Q-learning update
        best_next_action = np.argmax(Q[next_state-1, :]) if not done else 0
        # Explanation:
        #   next_state-1 because after stepping, state was incremented in env.
        #   If done = True, there's no next state to update from, so we take 0 by default.
        td_target = reward + (0 if done else (GAMMA * Q[next_state, best_next_action]))
        td_error = td_target - Q[state, action_idx]
        Q[state, action_idx] += ALPHA * td_error

        state = next_state
        total_reward += reward

        if done:
            break

    episode_rewards.append(total_reward)

    # Decay epsilon
    epsilon = max(EPSILON_END, epsilon * EPSILON_DECAY)

# -----------------------------
# 4. Visualization
# -----------------------------
# Let's plot the moving average of the episode rewards to see training progress.
window_size = 50
moving_avg = []
for i in range(len(episode_rewards)):
    start = max(0, i - window_size + 1)
    moving_avg.append(np.mean(episode_rewards[start:i+1]))

plt.figure(figsize=(10, 6))
plt.plot(episode_rewards, alpha=0.4, label='Episode Reward')
plt.plot(moving_avg, color='red', label=f'{window_size}-episode moving average')
plt.title("Q-learning for Dynamic Pricing: Training Rewards")
plt.xlabel("Episode")
plt.ylabel("Total Revenue per Episode")
plt.legend()
plt.show()

# -----------------------------
# 5. Inspect learned policy
# -----------------------------
# We can examine the best action (price) per state/time step after training.
optimal_policy = {}
for t in range(T_MAX):
    best_action_idx = np.argmax(Q[t, :])
    best_price = PRICE_SET[best_action_idx]
    optimal_policy[t] = best_price

print("Learned Optimal Policy (time step -> price):")
for t in range(T_MAX):
    print(f"  t={t}: price={optimal_policy[t]}")
```

```
import numpy as np
import matplotlib.pyplot as plt
import random

# -----------------------------
# 1. Define the Environment
# -----------------------------
class SeasonalPricingEnv:
    """
    A simple environment to simulate dynamic pricing with a seasonal demand.
    - State: time step t in [0, T_max-1]
    - Action: price chosen from a discrete set
    - Reward: price * demand
    - Demand is a function of time and price, plus random noise.
    """
    def __init__(self, T_max=10, price_set=[5, 6, 7, 8, 9]):
        self.T_max = T_max
        self.price_set = price_set
        self.num_actions = len(price_set)
        self.state = 0  # time index
        self.done = False

    def reset(self):
        """
        Reset the environment for a new episode.
        Returns the initial state.
        """
        self.state = 0
        self.done = False
        return self.state

    def step(self, action_idx):
        """
        Take one step in the environment given an action index.

        action_idx: index in self.price_set.
        Returns:
          next_state, reward, done
        """
        price = self.price_set[action_idx]

        # Simulate demand based on "seasonality"
        # For simplicity, let's split T_max into two "seasons":
        #   Early season (t < T_max/2): higher base demand
        #   Late season (t >= T_max/2): lower base demand
        if self.state < self.T_max // 2:
            base_demand = 100  # high-demand season
        else:
            base_demand = 60   # low-demand season

        # Demand decreases with price, plus Poisson noise
        demand_mean = max(base_demand - 5 * price, 0)  # keep at least 0
        demand = np.random.poisson(lam=demand_mean)

        # Revenue = price * demand
        reward = price * demand

        # Advance time
        self.state += 1
        if self.state >= self.T_max:
            self.done = True

        return self.state, reward, self.done

# -----------------------------
# 2. Q-learning Setup
# -----------------------------
NUM_EPISODES = 2000      # Number of training episodes
T_MAX = 10               # Steps in each episode
PRICE_SET = [5, 6, 7, 8, 9]
ALPHA = 0.1              # Learning rate
GAMMA = 0.95             # Discount factor
EPSILON_START = 1.0      # Initial exploration probability
EPSILON_END = 0.01       # Final exploration probability
EPSILON_DECAY = 0.999    # Exponential decay rate

# Create environment
env = SeasonalPricingEnv(T_max=T_MAX, price_set=PRICE_SET)

# Q-table shape: [T_max, number_of_actions]
Q = np.zeros((T_MAX, env.num_actions))

# Tracking total reward per episode
episode_rewards = []

# -----------------------------
# 3. Q-learning Training Loop
# -----------------------------
epsilon = EPSILON_START

for episode in range(NUM_EPISODES):
    state = env.reset()
    total_reward = 0

    for t in range(T_MAX):
        # Epsilon-greedy action selection
        if random.random() < epsilon:
            action_idx = np.random.randint(env.num_actions)
        else:
            action_idx = np.argmax(Q[state, :])

        next_state, reward, done = env.step(action_idx)

        # Q-learning update
        if not done:
            best_next_action = np.argmax(Q[next_state, :])
        else:
            best_next_action = 0  # No next state if done

        td_target = reward + (0 if done else GAMMA * Q[next_state, best_next_action])
        td_error = td_target - Q[state, action_idx]
        Q[state, action_idx] += ALPHA * td_error

        state = next_state
        total_reward += reward

        if done:
            break

    episode_rewards.append(total_reward)
    # Decay epsilon
    epsilon = max(EPSILON_END, epsilon * EPSILON_DECAY)

# -----------------------------
# 4. Visualize Training
# -----------------------------
# Let's plot the moving average of the episode rewards.
window_size = 50
moving_avg = []
for i in range(len(episode_rewards)):
    start = max(0, i - window_size + 1)
    moving_avg.append(np.mean(episode_rewards[start:i+1]))

plt.figure(figsize=(10, 4))
plt.plot(episode_rewards, alpha=0.4, label='Episode Reward')
plt.plot(moving_avg, color='red', label=f'{window_size}-ep moving avg')
plt.title("Training Rewards over Episodes")
plt.xlabel("Episode")
plt.ylabel("Total Revenue")
plt.legend()
plt.show()

# -----------------------------
# 5. Inspect Learned Policy
# -----------------------------
optimal_policy = {}
for t in range(T_MAX):
    best_action_idx = np.argmax(Q[t, :])
    best_price = PRICE_SET[best_action_idx]
    optimal_policy[t] = best_price

print("Learned Optimal Policy (time step -> price):")
for t in range(T_MAX):
    print(f"  t={t}: price={optimal_policy[t]}")

# -----------------------------
# 6. Visualize Price & Demand over Time
#    using the learned (greedy) policy
# -----------------------------
# We will run a few "test" episodes to see how price and demand change at each time step.
num_test_episodes = 20

all_prices = np.zeros((num_test_episodes, T_MAX))
all_demands = np.zeros((num_test_episodes, T_MAX))

for e in range(num_test_episodes):
    state = env.reset()
    for t in range(T_MAX):
        # Greedy action selection based on trained Q
        action_idx = np.argmax(Q[state, :])
        price = PRICE_SET[action_idx]

        # Step in the environment
        next_state, reward, done = env.step(action_idx)

        # Because reward = price * demand, we can recover demand:
        demand = 0
        if price > 0:
            demand = reward / price

        all_prices[e, t] = price
        all_demands[e, t] = demand

        state = next_state
        if done:
            break

# Compute average price and average demand over time
avg_price_over_time = np.mean(all_prices, axis=0)
avg_demand_over_time = np.mean(all_demands, axis=0)

# --- 6a. Plot side-by-side subplots ---
fig, axes = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

axes[0].plot(range(T_MAX), avg_price_over_time, marker='o', color='blue')
axes[0].set_ylabel("Price")
axes[0].set_title("Average Price over Time (Final Policy)")

axes[1].plot(range(T_MAX), avg_demand_over_time, marker='o', color='green')
axes[1].set_ylabel("Demand")
axes[1].set_xlabel("Time Step")
axes[1].set_title("Average Demand over Time (Final Policy)")

plt.tight_layout()
plt.show()

# --- 6b. (Optional) Plot on same subplot with dual y-axis ---
fig, ax1 = plt.subplots(figsize=(8, 4))

color1 = 'tab:blue'
ax1.set_xlabel("Time Step")
ax1.set_ylabel("Price", color=color1)
ax1.plot(range(T_MAX), avg_price_over_time, marker='o', color=color1, label='Price')
ax1.tick_params(axis='y', labelcolor=color1)

ax2 = ax1.twinx()  # second y-axis
color2 = 'tab:green'
ax2.set_ylabel("Demand", color=color2)
ax2.plot(range(T_MAX), avg_demand_over_time, marker='s', color=color2, label='Demand')
ax2.tick_params(axis='y', labelcolor=color2)

plt.title("Average Price & Demand Over Time (Final Policy)")
fig.tight_layout()
plt.show()
```
